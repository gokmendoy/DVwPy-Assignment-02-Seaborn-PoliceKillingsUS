## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill.head()
```

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

## 2: Plot 12 most dangerous cities

```python
kill.head()
```

```python
kill.city.value_counts().sort_index(ascending=True).sort_values(ascending=False).head(12)
```

```python
x = kill.city.value_counts().sort_index(ascending=True).sort_values(ascending=False).head(12).index
y = kill.city.value_counts().sort_index(ascending=True).sort_values(ascending=False).head(12).values

sns.barplot(data=kill, x= x, y= y);
plt.xticks(rotation=60)
plt.title('Most dangerous cities', color='blue')
```

```python
# with seaborn barplot

```

```python
kill.city.value_counts().head(12)
```

```python
frequency = kill.city.value_counts().sort_index(ascending=True).sort_values(ascending=False).head(12).index

sns.countplot(data=kill, x = 'city', order=frequency );
plt.xticks(rotation=60)
plt.title('Most dangerous cities', color='blue')
```

```python
# with seaborn countplot
```
